```python
# Clean categorical codes
# EDUCATION: valid {1,2,3,4}. Map {0,5,6} -> 4 ("others")
if 'education' in dfc.columns:
    dfc['education'] = pd.to_numeric(dfc['education'], errors='coerce').fillna(4).astype(int)
    dfc['education'] = dfc['education'].replace({0:4, 5:4, 6:4})

# MARRIAGE: valid {1,2,3}. Map 0 -> 3 ("others")
if 'marriage' in dfc.columns:
    dfc['marriage'] = pd.to_numeric(dfc['marriage'], errors='coerce').fillna(3).astype(int)
    dfc['marriage'] = dfc['marriage'].replace({0:3})

# SEX should be {1,2}
if 'sex' in dfc.columns:
    dfc['sex'] = pd.to_numeric(dfc['sex'], errors='coerce').fillna(2).astype(int)

# PAY_* are ordinal delinquency codes; clip to [-1, 9]
pay_cols = sorted([c for c in dfc.columns if c.startswith('pay_')],
                  key=lambda x: int(x.split('_')[1]) if x.split('_')[1].isdigit() else 999)
for c in pay_cols:
    dfc[c] = pd.to_numeric(dfc[c], errors='coerce').fillna(0).astype(int)
    dfc[c] = dfc[c].clip(lower=-1, upper=9)
```

```python
# Feature engineering
# Utilization per month: bill_amtk / limit_bal
if 'limit_bal' in dfc.columns:
    for k in range(1, 7):
        b = f'bill_amt{k}'
        if b in dfc.columns:
            dfc[f'utilization_{k}'] = dfc[b] / dfc['limit_bal'].replace(0, np.nan)
            dfc[f'utilization_{k}'] = dfc[f'utilization_{k}'].fillna(0.0)

# Payment ratio per month: pay_amtk / |bill_amtk|
for k in range(1, 7):
    p = f'pay_amt{k}'
    b = f'bill_amt{k}'
    if p in dfc.columns and b in dfc.columns:
        denom = (dfc[b].abs() + 1e-6)  # safe divide
        dfc[f'pay_ratio_{k}'] = dfc[p] / denom

# Aggregates that capture recent behavior
util_cols = [c for c in dfc.columns if c.startswith('utilization_')]
payr_cols = [c for c in dfc.columns if c.startswith('pay_ratio_')]

if util_cols:
    dfc['utilization_mean'] = dfc[util_cols].mean(axis=1)
    dfc['utilization_recent_mean'] = dfc[[c for c in util_cols if c.endswith('1') or c.endswith('2')]].mean(axis=1)

if payr_cols:
    dfc['pay_ratio_mean'] = dfc[payr_cols].mean(axis=1)
    dfc['pay_ratio_recent_mean'] = dfc[[c for c in payr_cols if c.endswith('1') or c.endswith('2')]].mean(axis=1)
```

```python
def choose_threshold_cost(y_true, y_score, cost_fp=1.0, cost_fn=5.0):
    thresholds = np.linspace(0.05, 0.95, 181)
    best = None
    for t in thresholds:
        y_pred = (y_score >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
        cost = cost_fp*fp + cost_fn*fn
        if (best is None) or (cost < best[0]):
            best = (cost, t, tp, fp, fn, tn)
    return best

best_cost, best_t, tp, fp, fn, tn = choose_threshold_cost(y_val, proba_cal, cost_fp=1.0, cost_fn=5.0)
print(f"Best threshold = {best_t:.2f}  |  Min expected cost = {best_cost}")
print(f"Confusion @t: TP={tp}, FP={fp}, FN={fn}, TN={tn}")

y_pred_best = (proba_cal >= best_t).astype(int)
print("\nClassification report @ best threshold:")
print(classification_report(y_val, y_pred_best, digits=3))
```

```python
# Build (N, 6, F) sequences for GRU

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Define the monthly feature columns
def month_cols(k):
    # k: 1..6
    b = f"bill_amt{k}"
    p = f"pay_amt{k}"
    u = f"utilization_{k}"
    r = f"pay_ratio_{k}"
    # delinquency code: month1 uses pay_0, months 2..6 use pay_2..pay_6
    if k == 1:
        d = "pay_0"
    else:
        d = f"pay_{k if k>=2 else 2}"  # k=2pay_2 ... k=6pay_6
    return [b, p, d, u, r]

seq_cols = [c for k in range(1,7) for c in month_cols(k)]
seq_cols = [c for c in seq_cols if c in df.columns]  # keep only existing

# Helper to stack months into (N,6,F)
def make_sequences(frame):
    # frame: DataFrame indexed to the rows we want (e.g., df.loc[X_train.index])
    # Build per-month tensors and stack on time axis
    month_tensors = []
    for k in range(1,7):
        cols = [c for c in month_cols(k) if c in frame.columns]
        M = frame[cols].copy()
        # ensure all numeric
        for c in cols:
            M[c] = pd.to_numeric(M[c], errors='coerce').fillna(0.0)
        month_tensors.append(M.values)
    # shape per month: (N, F_k). We need same F each month; pad if needed.
    maxF = max(m.shape[1] for m in month_tensors)
    padded = []
    for M in month_tensors:
        if M.shape[1] < maxF:
            pad = np.zeros((M.shape[0], maxF - M.shape[1]), dtype=float)
            M = np.hstack([M, pad])
        padded.append(M)
    # stack months into time axis: list length 6 -> (6, N, F) -> transpose to (N,6,F)
    X_seq = np.stack(padded, axis=1)  # (N, 6, F)
    return X_seq

# Build train/val sequences matching your existing split
X_train_seq = make_sequences(df.loc[X_train.index])
X_val_seq   = make_sequences(df.loc[X_val.index])

# Standardize features across the training set (flatten -> scale -> reshape)
Ntr, T, F = X_train_seq.shape
scaler = StandardScaler()
X_train_seq_2d = X_train_seq.reshape(Ntr, T*F)
X_train_seq_2d = scaler.fit_transform(X_train_seq_2d)
X_train_seq = X_train_seq_2d.reshape(Ntr, T, F)

Nva = X_val_seq.shape[0]
X_val_seq_2d = X_val_seq.reshape(Nva, T*F)
X_val_seq_2d = scaler.transform(X_val_seq_2d)
X_val_seq = X_val_seq_2d.reshape(Nva, T, F)

print("Sequence shapes  train:", X_train_seq.shape, " val:", X_val_seq.shape)
```

```python
# Train a compact GRU model

import numpy as np, tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Class weights (same idea as scale_pos_weight)
pos = int(y_train.sum()); neg = int(len(y_train) - pos)
class_weight = {0: 1.0, 1: max(1.0, neg/pos)}

# Model
tf.keras.backend.clear_session()
model = keras.Sequential([
    layers.Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2])),
    layers.GRU(32, return_sequences=False),
    layers.Dense(16, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=[keras.metrics.AUC(name='auc_roc'),
             keras.metrics.AUC(name='auc_pr', curve='PR')]
)

cb = [
    keras.callbacks.EarlyStopping(monitor='val_auc_pr', mode='max',
                                  patience=5, restore_best_weights=True)
]

hist = model.fit(
    X_train_seq, y_train.values,
    validation_data=(X_val_seq, y_val.values),
    epochs=30, batch_size=256,
    class_weight=class_weight,
    callbacks=cb, verbose=1
)

# Predicted probabilities on validation
proba_gru = model.predict(X_val_seq, batch_size=512).ravel()
```

```python
import numpy as np, pandas as pd
from sklearn.metrics import confusion_matrix, average_precision_score, roc_auc_score

y_score = proba_cal if 'proba_cal' in globals() else proba_xgb
ratios = [(3,1),(5,1),(7,1)]

def pick_t(y_true, y_score, C_FN, C_FP):
    ths = np.linspace(0.01,0.99,199); best=(None,None)
    for t in ths:
        y_pred = (y_score>=t).astype(int)
        tn,fp,fn,tp = confusion_matrix(y_true,y_pred).ravel()
        cost = C_FP*fp + C_FN*fn
        if best[0] is None or cost<best[0]: best=(cost,(t,tp,fp,fn,tn))
    return best

rows=[]
for C_FN,C_FP in ratios:
    cost,(t,tp,fp,fn,tn)=pick_t(y_val.values,y_score,C_FN,C_FP)
    prec = tp/(tp+fp+1e-9); rec = tp/(tp+fn+1e-9)
    rows.append([f"{C_FN}:{C_FP}", round(t,2), round(prec,3), round(rec,3), int(cost), tp, fp, fn, tn])

print(pd.DataFrame(rows, columns=["FN:FP","t*","Precision","Recall","Cost","TP","FP","FN","TN"]))
print(f"\nROC-AUC={roc_auc_score(y_val,y_score):.3f} | PR-AUC={average_precision_score(y_val,y_score):.3f}")
```
