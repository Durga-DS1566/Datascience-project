```python
# categorical summaries with default rate
def cat_summary(df, col, target='default_payment_next_month'):
    g = df.groupby(col)[target].agg(['count','mean']).rename(columns={'mean':'default_rate'})
    g['default_rate'] = (g['default_rate']*100).round(2)
    return g.sort_index()

for col in ['sex','education','marriage']:
    if col in df.columns:
        display(cat_summary(df, col))
```

```python
dfc = df.copy()
```

```python
# Clean categorical codes
# EDUCATION: valid {1,2,3,4}. Map {0,5,6} -> 4 ("others")
if 'education' in dfc.columns:
    dfc['education'] = pd.to_numeric(dfc['education'], errors='coerce').fillna(4).astype(int)
    dfc['education'] = dfc['education'].replace({0:4, 5:4, 6:4})

# MARRIAGE: valid {1,2,3}. Map 0 -> 3 ("others")
if 'marriage' in dfc.columns:
    dfc['marriage'] = pd.to_numeric(dfc['marriage'], errors='coerce').fillna(3).astype(int)
    dfc['marriage'] = dfc['marriage'].replace({0:3})

# SEX should be {1,2}
if 'sex' in dfc.columns:
    dfc['sex'] = pd.to_numeric(dfc['sex'], errors='coerce').fillna(2).astype(int)

# PAY_* are ordinal delinquency codes; clip to [-1, 9]
pay_cols = sorted([c for c in dfc.columns if c.startswith('pay_')],
                  key=lambda x: int(x.split('_')[1]) if x.split('_')[1].isdigit() else 999)
for c in pay_cols:
    dfc[c] = pd.to_numeric(dfc[c], errors='coerce').fillna(0).astype(int)
    dfc[c] = dfc[c].clip(lower=-1, upper=9)
```

```python
# Feature engineering
# Utilization per month: bill_amtk / limit_bal
if 'limit_bal' in dfc.columns:
    for k in range(1, 7):
        b = f'bill_amt{k}'
        if b in dfc.columns:
            dfc[f'utilization_{k}'] = dfc[b] / dfc['limit_bal'].replace(0, np.nan)
            dfc[f'utilization_{k}'] = dfc[f'utilization_{k}'].fillna(0.0)

# Payment ratio per month: pay_amtk / |bill_amtk|
for k in range(1, 7):
    p = f'pay_amt{k}'
    b = f'bill_amt{k}'
    if p in dfc.columns and b in dfc.columns:
        denom = (dfc[b].abs() + 1e-6)  # safe divide
        dfc[f'pay_ratio_{k}'] = dfc[p] / denom

# Aggregates that capture recent behavior
util_cols = [c for c in dfc.columns if c.startswith('utilization_')]
payr_cols = [c for c in dfc.columns if c.startswith('pay_ratio_')]

if util_cols:
    dfc['utilization_mean'] = dfc[util_cols].mean(axis=1)
    dfc['utilization_recent_mean'] = dfc[[c for c in util_cols if c.endswith('1') or c.endswith('2')]].mean(axis=1)

if payr_cols:
    dfc['pay_ratio_mean'] = dfc[payr_cols].mean(axis=1)
    dfc['pay_ratio_recent_mean'] = dfc[[c for c in payr_cols if c.endswith('1') or c.endswith('2')]].mean(axis=1)
```

```python
# Quick preview & hand-off
target = 'default_payment_next_month'
new_cols = [c for c in ['utilization_recent_mean','pay_ratio_recent_mean',
                        'utilization_1','utilization_2','pay_ratio_1','pay_ratio_2'] if c in dfc.columns]

print("New feature samples:")
display(dfc[new_cols].head(5))

print("\nPAY_* columns clipped to range [-1,9]:")
print({c: (int(dfc[c].min()), int(dfc[c].max())) for c in pay_cols})

print("\nShape after feature engineering:", dfc.shape)

# Replace df with cleaned+featured version for next steps
df = dfc
```

```python
df.head()
```

```python
target = 'default_payment_next_month'
X = df.drop(columns=[target])
y = df[target].astype(int)

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
X_train.shape, X_val.shape
```

```python
# numeric columns only (all are numeric in this dataset)
num_cols = X_train.columns.tolist()
num_cols
```

```python
# Logistic Regression
preprocess = ColumnTransformer([('num', StandardScaler(), num_cols)], remainder='drop')
log_reg = Pipeline([
    ('prep', preprocess),
    ('clf', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=RANDOM_STATE))
])
log_reg.fit(X_train, y_train)
proba_lr = log_reg.predict_proba(X_val)[:,1]
print("Logistic:")
print(f"ROC-AUC={roc_auc_score(y_val, proba_lr):.3f}")
print(f"PR-AUC={average_precision_score(y_val, proba_lr):.3f}")
```

```python
# XGBoost (class imbalance handled via scale_pos_weight)
pos_weight = (len(y_train) - y_train.sum()) / y_train.sum()
xgb = XGBClassifier(
    n_estimators=400, max_depth=4, learning_rate=0.05,
    subsample=0.9, colsample_bytree=0.8,
    reg_alpha=0.0, reg_lambda=1.0,
    random_state=RANDOM_STATE, n_jobs=-1,
    scale_pos_weight=pos_weight
)
xgb.fit(X_train, y_train)
proba_xgb = xgb.predict_proba(X_val)[:,1]
print(f"XGBoost   ROC-AUC={roc_auc_score(y_val, proba_xgb):.3f} | PR-AUC={average_precision_score(y_val, proba_xgb):.3f}")
```

```python
cal = CalibratedClassifierCV(xgb, method='isotonic', cv=3)
cal.fit(X_train, y_train)
proba_cal = cal.predict_proba(X_val)[:,1]

print(f"Calibrated XGB ROC-AUC={roc_auc_score(y_val, proba_cal):.3f} | PR-AUC={average_precision_score(y_val, proba_cal):.3f}")
print(f"Brier score (lower is better) = {brier_score_loss(y_val, proba_cal):.4f}")

# quick reliability curve (optional visualization)
from sklearn.calibration import CalibrationDisplay
disp = CalibrationDisplay.from_predictions(y_val, proba_cal, n_bins=10)
plt.title("Calibration: Calibrated XGB")
plt.show()
```

```python
def choose_threshold_cost(y_true, y_score, cost_fp=1.0, cost_fn=5.0):
    thresholds = np.linspace(0.05, 0.95, 181)
    best = None
    for t in thresholds:
        y_pred = (y_score >= t).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
        cost = cost_fp*fp + cost_fn*fn
        if (best is None) or (cost < best[0]):
            best = (cost, t, tp, fp, fn, tn)
    return best

best_cost, best_t, tp, fp, fn, tn = choose_threshold_cost(y_val, proba_cal, cost_fp=1.0, cost_fn=5.0)
print(f"Best threshold = {best_t:.2f}  |  Min expected cost = {best_cost}")
print(f"Confusion @t: TP={tp}, FP={fp}, FN={fn}, TN={tn}")

y_pred_best = (proba_cal >= best_t).astype(int)
print("\nClassification report @ best threshold:")
print(classification_report(y_val, y_pred_best, digits=3))
```
